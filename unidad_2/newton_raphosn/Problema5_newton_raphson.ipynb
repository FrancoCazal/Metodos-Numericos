Una lata en forma de un cilindro circular recto se construye para contener $1000 \text{cm}^3$. Las partes superior e inferior de la lata deben tener un radio de $0.25 \text{cm}$ más que el radio de la lata, de tal forma que el exceso se pueda usar para formar un sello con la parte lateral. La hoja de material que se forma dentro de la parte lateral de la lata también debe ser $0.25 \text{cm}$ más grande que la circunferencia de la lata, de tal forma que se pueda formar un sello. Encuentre, dentro de $10^{-4}$, la cantidad mínima de material necesario para construir la lata.

### Planteamiento del problema
La capacidad volumetrica del cilindro esta dada por:
$$1000 = \pi r^2 h$$
Entonces $h = \frac{1000}{\pi r^2}$

Ademas, para el area total de la lata es:
$$
\begin{align*}
A &= 2A_{tapa} + A_{lateral} \\
&= 2\pi(r+0.25)^2 + 2\pi(r+0.25)h \\
&= 2\pi(r+0.25)^2 + 2\pi(r+0.25)\frac{1000}{\pi r^2} \\
\end{align*}
$$
Luego, el minimo de capacidad volumetrica verifica $f(r) = A'(r) = 0$

```python
import numpy as np
from jax import grad # permite realiza derivadas

# Definicion de funcion f(x) = 0
def A(r):
    return 2*np.pi*(r+0.25)**2 + 2*np.pi*(r+0.25)*(1000/(np.pi*r**2))

f = grad(A)

# Condiciones iniciales
p0 = 1.0
tol = 1e-4
n = 10

# %%  RUTINA007##########################
df = grad(f) # jax reconoce automaticamente que debe derivar respecto al parámetro de f definido previamente Ej: def f(r) # deriva respecto a r
for i in range(n):
 p=p0-f(p0)/df(p0)
 err=np.abs(p-p0)
 relerr=err/np.abs(p)
 if tol>err or tol>relerr or tol>np.abs(f(p)):
   break
 p0=p

print("Verificacion de convergencia: ")
print("error absoluto: ", err)
print("error relativo: ", relerr)
print("valor f(p): ", np.abs(f(p)))
print("numero de iteraciones: ",i+1)

print("-------------")
print("a) En valor de r que minimiza A es: ", np.around(p,6))
print("b) La cantidad minima de materia necesario es: ",np.around(A(p),6))



```
