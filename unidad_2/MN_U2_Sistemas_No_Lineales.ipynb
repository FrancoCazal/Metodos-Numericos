## 1.1. Caso Unidimensional

Usualmente los sistemas de ecuaciones no lineales (no reducibles a como $Ax=b$) no presentan solución analítica o resulta inaccesible por el coste computacional que conlleva, por ello se recurre a métodos iterativos (como anteriormente vimos con el Método de Gauss-Seidel y Gauss-Jacobi, que son métodos diseñados para sistemas lineales) que se aproximan a la solución mediante iteraciones sucesivas de un proceso que converge a la solución. De esta forma, la aproximación se *refina* sucesivamente hasta alcanzar la precisión deseada (*tolerancia*).

En particular en el caso Unidimensional, es decir sea $(f:[a,b]\to\mathbb{R}$ una función continua. El objetivo es encontrar $x$ tal que
$$
f(x)=0.
$$


### a. Método de Bisección
El método de la bisección se basa en reducir sistemáticamente el intervalo en el que se encuentra la raíz partiendo de un intervalo con cambio de signo de la función.

Para garantizar la existencia (y en ciertos casos la unicidad) de la raíz se utiliza el siguiente resultado:

**Teorema:** Sea $f:[a,b]\to\mathbb{R}$ continua. Si $f(a)f(b)<0$, entonces existe al menos un punto $\bar{x} \in (a,b)$ tal que $f(\bar{x})=0$. Si además $f'(\bar{x}) \in (a,b)$ y no cambia de signo, entonces dicha raíz es única.

El procedimiento se inicia definiendo $a_0=a$ y $b_0=b$. En cada iteración se calcula el punto medio:
$$
c_{k+1}=\frac{a_k+b_k}{2},
$$
y se evalúa $f(c_{k+1})$.

Si $f(c_{k+1})=0$, se ha hallado la solución; en caso contrario, se determina en qué subintervalo se mantiene el cambio de signo, es decir, se actualiza el intervalo según:
$$
\begin{cases}
a_{k+1}=a_k,\quad b_{k+1}=c_{k+1}, & \text{si } f(a_k)f(c_{k+1})<0,\\[2mm]
a_{k+1}=c_{k+1},\quad b_{k+1}=b_k, & \text{si } f(c_{k+1})f(b_k)<0.
\end{cases}
$$
El proceso se repite hasta que la longitud del intervalo $|b_k-a_k|$ sea menor que la tolerancia deseada.

Una interpretación de este método es que "acorrala" a la raíz reduciendo sucesivamente el rango en que se encuentra en un factor de 2.

### b. Método de la Falsa Posición
Este método es similar al método de bisección, también utiliza un intervalo $[a_k,b_k]$ con cambio de signo $f(a_k)f(b_k)<0$, pero en lugar de tomar el punto medio, obtiene la intersección de la recta secante que une los puntos $A(a_k, f(a_k)$ y $B(b_k, f(b_k))$ con el eje $x$.

La fórmula iterativa resultante es:
$$
c_{k+1} = \frac{a_k\,|f(b_k)|+b_k\,|f(a_k)|}{|f(a_k)|+|f(b_k)|} = \frac{a_kf(b_k)-b_kf(a_k)}{f(b_k)-f(a_k)}.
$$
Una vez calculado $c_{k+1}$, se actualiza el intervalo de forma similar a la bisección, seleccionando el subintervalo en el que se mantiene el cambio de signo.


### c. Método del Punto Fijo
El método del punto fijo consiste en reescribir la ecuación $f(x)=0$ en la forma
$x=g(x)$ (no siempre es posible) y luego construir la secuencia iterativa:
$$
x_{k+1}=g(x_k).
$$
Para asegurar la convergencia es necesario que $g$ sea contractiva, es decir, que exista una constante $L$ con $0 \leq L<1$ tal que
$$
|g'(x)|\leq L,\quad \forall\, x\in [a,b].
$$

### d. Método de Newton-Raphson
El método de Newton-Raphson se basa en aproximar $f$ mediante su expansión en serie de Taylor en torno a $x_k$. Despreciando los términos de orden superior se obtiene la secuencia iterativa
$$
x_{k+1}=x_k-\frac{f(x_k)}{f'(x_k)}.
$$
La iterpretación geometrica del metodo es la de una recta tangente a la curva de $f$ en el punto $x_k$ en $(x,f(x))$ que corta el eje $x$ dicha intersección es el punto $x_{k+1}$.

### e. Método de la Secante
El método de la secante es una version mas simple del Método de Newton-Raphson que evita el cálculo directo de la derivada de $f$ aproximándola mediante una diferencia finita. La iteración se define por:
$$
x_{k+1}=x_k-f(x_k)\frac{x_k-x_{k-1}}{f(x_k)-f(x_{k-1})}.
$$
Se utilizan dos aproximaciones previas para calcular el siguiente valor sin necesidad de conocer $f'(x)$.

### Ejemplo `RUTINA4`

Resolver $x^{3/2} + x - 3 = 0$ con el M. de Biseccion bajo:
- $ tol = 10^{-6}$
- 6 decimales de precision en la respuesta
- un número máximo de iteraciones de 3
- Usar todos los decimales para los calculos

````python
import numpy as np
from jax import grad # permite realiza derivadas
import jax.numpy as jnp # es el numpy integrado de jax

# Resolver x^(3/2) + x = 3
# Definimos la f(x) = 0 como f(x) = x^(3/2) + x - 3 = 0

# DEFINCION DE f(x):
def f(x):
    return np.pow(x,3/2) + x - 3

# PARAMETROS DE INICIALIZACION
tol = 1e-6           # tolerancia
n = 3                # número máximo de iteraciones
a = 1.0                # limite inferior
b = 2.0                # limite superior
c0 = a               # estimacion inicial

################RUTINA004##########################
for i in range(n):
 c=(a+b)/2
 err=np.abs(c0-c)
 relerr=np.abs(err/c)
 if tol>err or tol>relerr or tol>np.abs(f(c)):
   break
 else:
   if f(a)*f(c)<0:
     b=c
   else:
     a=c
 c0=c

print("Verificacion de convergencia: ")
print("error absoluto: ", err)
print("error relativo: ", relerr)
print("valor f(p): ", np.abs(f(c)))
print("numero de iteraciones: ",i+1)

print("-----Respuesta-----")
print("Solucion: ",np.around(c,6))
````

El caso multidimensional es similar sin embargo se tiene, $F(x_1,x_2,x_3...x_n) = 0_m: R^n \to R^m$. En terminos mas especificos si es resolver $m$ ecuaciones $f_i(x_1,x_2...x_n) = 0$ al mismo tiempo o bien
$$
f_1(x_1, x_2, x_3, ... x_n) = 0 \\
f_2(x_1, x_2, x_3, ... x_n) = 0 \\
...  \\
f_m(x_1, x_2, x_3, ... x_n) = 0 \\
$$
Definiendo $X = (x1,x2,... x_n)$ y $F = (f_1,f_2,f_3...f_m)$ equivale a $F(X) = 0$, al igual que el caso unidimensional la aproximación se _refina_ sucesivamente partiendo de una solucion inicial ($p_0$) hasta alcanzar la precisión/tolerancia deseada (*$tol$*) o alcanzar un numero limite de iteraciones ($m$)

El mapa iterativo del método de Newton es:
$$
    X_{k+1} = X_k - J(X_k)^{-1} F(X_k).
$$
donde $J(X_k)$ es la matriz Jacobiana de $F$, dada por:
\begin{equation}
    J(X) = \begin{bmatrix} \nabla f_1 \\ \nabla f_2 \\ \vdots \\ \nabla f_n \end{bmatrix} = \begin{bmatrix} \frac{\partial f_1}{\partial x_1} & \cdots & \frac{\partial f_1}{\partial x_m} \\
    \frac{\partial f_2}{\partial x_1} & \cdots & \frac{\partial f_2}{\partial x_m} \\
    \vdots & \ddots & \vdots \\
    \frac{\partial f_n}{\partial x_1} & \cdots & \frac{\partial f_n}{\partial x_m} \end{bmatrix}.
\end{equation}

Desde una perspectiva de optimización de una funcion escalar $h(X)$ con $m$ restricciones $f_i(X) = 0$, el multiplicador de Lagrange es
$$
    L(X, \lambda_1, \lambda_2, ...) = h(X) + \sum_{i=1}^m \lambda_i f_i(X)
$$
Entonces las ecuaciones que constituyen el punto maximo o minimo es:
$$
    \nabla L(X, \lambda_1, \lambda_2, ...) = \begin{bmatrix} 0 \\ \vdots \\ 0 \end{bmatrix} = 0
$$

Que constituye un sistema de ecuaciones no lineales resoluble con el metodo de Newton.

**Problema 2:** Un estudiante analiza las siguientes alternativas para llegar a la facultad partiendo de su casa en el menor tiempo posible:

- **Opción 1:** recorrer todo el trayecto en bicicleta.  
- **Opción 2:** recorrer una parte en bicicleta y luego en colectivo.  
- **Opción 3:** recorrer una parte en bicicleta y luego en automóvil.  

La distancia total a recorrer es de 5 $\text{(km)}$ y lo recorrido por cada medio de transporte en funcion del tiempo $t_i \text{(horas)}$ esta dado por:

$$
\begin{aligned}
\text{(bicicleta)} \quad & D_1(t_1) = \frac{1}{(t_1+1)^2} - t_1 + 6 \text{(km)}\\
\text{(colectivo)} \quad & D_2(t_2) = \frac{e^{t_2}}{t_2 - 3} \text{(km)}\\
\text{(automóvil)} \quad & D_3(t_3) = \frac{2 - t_3}{\sqrt{t_3^3 + 1}} + 4 \text{(km)}
\end{aligned}
$$

Relaciones entre tiempos: $t_1 = \frac{t_2}{2.6}, \qquad t_2 = 0.66\,t_3 \to t_1 = \frac{0.66t_3}{2.6}$

Determinar la mejor alternativa empleando el método de Newton-Raphson bajo.

- $ tol = 10^{-6}$
- 6 decimales de precision en la respuesta
- un número máximo de iteraciones de 10
- Usar todos los decimales para los calculos

_Planteamiento de la solución_

Definimos las funciones soluciones $f(t) = 0$

- Opción 1 (Solo bicicleta)
    $$
    f_1(t_1) = D_1(t_1) - 5 = 0
    $$

    $$
    f_1(t_1) = \frac{1}{(t_1+1)^2} - t_1 + 1 = 0
    $$

    Tiempo total: $T = t_1$

- Opción 2 (Bicicleta + Colectivo)
    $$
    f_2(t_1,t_2) = D_1(t_1) + D_2(t_2) - 5 = 0
    $$

    $$
    f_2(t_2) =
    \frac{1}{\left(\frac{t_2}{2.6}+1\right)^2}
    -\frac{t_2}{2.6}
    +1
    +\frac{e^{t_2}}{t_2-3}
    = 0
    $$

    Tiempo total: $T = \frac{t_2}{2.6} + t_2$


- Opción 3 (Bicicleta + Automóvil)
    $$
    f_3(t_1,t_3) = D_1(t_1) + D_3(t_3) - 5 = 0
    $$

    $$
    f_3(t_3) =
    \frac{1}{\left(\frac{0.66t_3}{2.6}+1\right)^2}
    -\frac{0.66t_3}{2.6}+
    \frac{2-t_3}
    {\sqrt{t_3^3+1}}
    +5
    = 0
    $$

Tiempo total: $T = \frac{0.66t_3}{2.6} + t_3$

Version 1: Definiendo D1, D2, y D3

Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa $m$ (lb) cae desde una altura $s_0$ (ft) y que la altura del objeto después de $t$ (s) es:

$$s(t)=s_0-\frac{mg}{k}t+\frac{m^2g}{k^2}(1-e^{-\frac{kt}{m}})$$

donde $g$ (ft/s²) es la gravedad y $k$ (lb.s/ft) representa el coeficiente de la resistencia del aire.

Sabiendo que de acuerdo a las pruebas realizadas se obtuvieron los siguientes resultados:

| Masa ($m$) | Tiempo ($t$) | Altura ($s$) |
| :---: | :---: | :---: |
| 100 | 0.5 | 35.8762 |
| 60 | 0.7 | 31.9207 |
| 20 | 0.1 | 39.835 |

Utilizando el método de Newton para resolver el sistema, con una tolerancia de $10^{-8}$, y un valor inicial unitario para la iteración inicial.

_Planteamiento_

Definiendo $f(s_0, g, k) = s_0 -\frac{mg}{k}t+\frac{m^2g}{k^2}(1-e^{-\frac{kt}{m}}) - s_t = 0$ para cada experimento de la tabla resulta:

$$
f_1(s_0, g, k) = s_0 -\frac{100g}{k}0.5+\frac{100^2g}{k^2}(1-e^{-\frac{0.5k}{100}}) - 35.8762 = 0
$$
$$
f_2(s_0, g, k) = s_0 -\frac{60g}{k}0.7+\frac{60^2g}{k^2}(1-e^{-\frac{0.7k}{60}}) - 31.9207 = 0
$$
$$
f_3(s_0, g, k) = s_0 -\frac{20g}{k}0.1+\frac{20^2g}{k^2}(1-e^{-\frac{0.1k}{20}}) - 39.835 = 0
$$
Lo anterior constituye un sistema de ecuaciones no lineal resoluble con el metodo de Newton-Raphson

````python
import jax.numpy as jnp
from jax import jacfwd

# Constantes

def sistema(P):
    s0, g, k = P
    f1 = s0 - (100*g/k)*0.5 + (jnp.pow(100,2)*g/jnp.pow(k,2))*(1-jnp.exp(-0.5*k/100)) - 35.8762
    f2 = s0 - (60*g/k)*0.7 + (jnp.pow(60,2)*g/jnp.pow(k,2))*(1-jnp.exp(-0.7*k/60)) - 31.9207
    f3 = s0 - (20*g/k)*0.1 + (jnp.pow(20,2)*g/jnp.pow(k,2))*(1-jnp.exp(-0.1*k/20)) - 39.835
    return jnp.array([f1, f2, f3], dtype = float)



jacob_sist = jacfwd(sistema)

# Parametros de inicializacion
P0 = jnp.array([1.0,1.0,1.0],dtype = float)
tol = 1e-8
m = 30

################RUTINA009##########################
for i in range(m):
    F = sistema(P0)
    J = jacob_sist(P0)
    deltaP = jnp.linalg.solve(J, -F)
    P = P0 + deltaP
    err = jnp.linalg.norm(deltaP)
    relerr = err / jnp.linalg.norm(P)
    f_norm = jnp.linalg.norm(F)
    if (err < tol) or (relerr < tol) or (f_norm < tol):
        P0 = P
        break
    P0 = P
print("Analisis de convergencia: ")
print("error absoluto: ", err)
print("error relativo: ", relerr)
print("F(P): ", jnp.linalg.norm(sistema(P)))
print("numero de iteraciones: ",i+1)

print("Respuesta:")
s0_res, g_res, k_res = P

print("posicion inicial: :",jnp.around(s0_res,6))
print("gravedad: ",jnp.around(g_res,6))
print("constante de rozamiento: ",jnp.around(k_res,6))

````

### Consideraciones para la Resolución

Se quiere construir un silo (incluir la base circular). Tiene forma de cilindro rematado por una semiesfera en el techo. El costo de construcción por unidad cuadrada del área superficial es de 24 dolares para la semiesfera, para la pared cilíndrica de 14 dolares y la base de 35 dolares por cada metro cuadrado. Determinar las dimensiones (radio de semiesfera y altura del cilindro) que se deben utilizar si se quiere minimizar los costos y el volumen fijo se pretende sea de 1000 metros cúbicos.
Utilizar el método de Newton directo con ayuda de un multiplicador de Lagrange de modo a minimizar el gasto. Utilizar como valor inicial al valor de 10 de todas las variables y una tolerancia de $10^{-3}$.

Determinar:
- Item a: Dimensiones del radio
- Item b: Dimensiones de la altura
- Item c: Valor absoluto del multiplicador de Lagrange
- Item d: Costo del Silo

_Planteamiento_

Para las Áreas de las componentes del silo:
$$
\begin{aligned}
A_p &= 2\pi r h \quad \text{(pared cilíndrica)} \\
A_t &= 2\pi r^2 \quad \text{(techo semiesférico)} \\
A_b &= \pi r^2 \quad \text{(base circular)}
\end{aligned}
$$

El costo total está dado por la suma de los costos de cada parte:
$$
\begin{aligned}
C(A_p, A_t, A_b) &= 14A_p + 24A_t + 35A_b \\
C(r, h) &= 14(2\pi r h) + 24(2\pi r^2) + 35(\pi r^2) \\
C(r, h) &= 28\pi r h + 83\pi r^2
\end{aligned}
$$

El volumen total se compone del cilindro y la semiesfera:
$$
\begin{aligned}
V_p &= \pi r^2 h \quad \text{(Interior de la pared cilíndrica)} \\
V_t &= \frac{2}{3}\pi r^3 \quad \text{(Interior del techo semiesférico)} \\
V_T &= \pi r^2 h + \frac{2}{3}\pi r^3 = 1000
\end{aligned}
$$

Definimos la función de Lagrange $L(r, h, \lambda)$ considerando la restricción de volumen:
$$L(r, h, \lambda) = (28\pi r h + 83\pi r^2) + \lambda \left( 1000 - \pi r^2 h - \frac{2}{3}\pi r^3 \right)$$

Para encontrar los puntos críticos, resolvemos:
$$
\begin{aligned}
\frac{\partial L}{\partial r} &= 0 \\
\frac{\partial L}{\partial h} &= 0 \\
\frac{\partial L}{\partial \lambda} &= 0
\end{aligned}
$$
Es decir $\nabla L = 0$ siendo $\nabla = \left( \frac{\partial }{\partial r}, \frac{\partial }{\partial h}, \frac{\partial }{\partial \lambda} \right)$.

````python
import jax.numpy as jnp
from jax import jacfwd
from jax import grad

# Definir las areas
def Ab(r):
    return jnp.pi*jnp.pow(r,2)

def Ac(r,h):
    return 2*jnp.pi*r*h

def As(r):
    return 2*jnp.pi*jnp.pow(r,2)

# Definimos los volumenes
def Vc(r,h):
    return jnp.pi*jnp.pow(r,2)*h

def Vs(r):
    return (2/3)*jnp.pi*jnp.pow(r,3)

# M. de Lagrange
def L(var):
    r,h,lm = var
    return (24*As(r)+14*Ac(r,h)+35*Ab(r)) + lm*(Vc(r,h) + Vs(r) - 1000)

sistema = grad(L)
jacob_sist = jacfwd(sistema)

# Parametros de inicializacion
P0 = jnp.array([10.0, 10.0, 10.0],dtype = float)
tol = 1e-3
m = 10

################RUTINA009##########################
for i in range(m):
    F = sistema(P0)
    J = jacob_sist(P0)
    deltaP = jnp.linalg.solve(J, -F)
    P = P0 + deltaP
    err = jnp.linalg.norm(deltaP)
    relerr = err / jnp.linalg.norm(P)
    f_norm = jnp.linalg.norm(F)
    if (err < tol) or (relerr < tol) or (f_norm < tol):
        P0 = P
        break
    P0 = P
````
