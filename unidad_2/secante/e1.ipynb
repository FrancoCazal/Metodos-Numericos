## Problema 1 - Solución con Método de la Secante
### Consideraciones para la Resolución

* **Utilizar seis decimales para los resultados.**
* **Utilizar todos los decimales para los calculos.**
* **No utilizar formato exponencial (notacion científica).**
* **No modificar los valores iniciales ni la tolerancia para los calculos.**
* **No es necesario realizar ninguna conversion de unidades.**

La velicidad $v$ de un paracaidista que cae está dada por:
$$
v = \frac{gm(1-e^{-\frac{ct}{m}})}{c}
$$
donde $g = 9.81 \text{ m/s}^2$. Para un paracaidista con coeficiente de arrastre $c = 15 \text{ kg/s}$, calcule la masa $m$ de modo que la velocidad sea $v = 35 \text{ m/s}$ en $t = 9 \text{ s}$. Utilice el método de la secante, en el intervalo de $59$ a $60$ para determinar $m$ a un nivel de tolerancia de $0.001$.
Determinar:
- Item a: El error absoluto cometido
- Item b: El error relativo cometido
- Item c: Cantidad de iteraciones
- Item d: El valor de masa $m$


```python
import numpy as np

# Constantes
g = 9.81
c = 15
v_value = 35
t = 9

# Definicion de funcion f(x) = 0 -> f(x) = v_value - v(x) = 0
def v(m):
    return (g*m*(1-np.exp((-c*t)/m)))/c

def f(x):
    return v_value - v(x)

# Parametros de inicializacion
x0=59
x1=60
x2=x1

n = 10
tol = 0.001

# %%  RUTINA008##########################
for i in range(n):
    x2=(x1*f(x0)-x0*f(x1))/(f(x0)-f(x1))
    err=np.abs(x2-x1)
    relerr=err/np.abs(x2)
    if tol>err or tol>relerr or tol>np.abs(f(x2)):
      break
    x0=x1
    x1=x2

print("Verificacion de convergencia:")
print("Error absoluto: ", err)
print("Error relativo: ", relerr)
print("Valor de f(x2): ", np.abs(f(x2)))
print("Numero de iteraciones: ", i+1)

print("------------------")
print("a) El error absoluto: ", np.around(err,6))
print("b) El error relativo: ", np.around(relerr,6))
print("c) Cantidad de iteraciones: ", i+1)
print("d) El valor de masa m: ", np.around(x2,6))
```
