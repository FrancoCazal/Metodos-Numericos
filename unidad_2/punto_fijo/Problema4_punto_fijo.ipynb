Dos escaleras se entrecruzan en un pasillo de ancho W. Cada escalera llega desde la base de una pared hasta algún punto en la pared opuesta. Las escaleras cruzan a una altura $H$ sobre el pavimento. Encuentre $W$ dado que las longitudes de las escaleras son $x_1 = 20 \text{ ft}$ y $x_2 = 30 \text{ ft}$ y que $H = 8 \text{ ft}$.

Utilizando el método de Punto fijo, determinar::
- Item a: Cantidad de iteraciones realizadas
- Item b: Valor de $W$
- Item c: Error cometido
- Item d: Error relativo

### Planteamiento del Problema
De la semejanza de triangulos, resultan las ecuaciones:
$$
\frac{H}{h_1}=\frac{l_2}{W} \text{ (1)}
$$
$$
\frac{H}{h_2}=\frac{l_1}{W} \text{ (2)}
$$

Sumando (1) y (2) resulta
$$
\frac{H}{h_1} + \frac{H}{h_2} = \frac{l_1 + l_2}{W}
$$
Pero como $ W = l_1 + l_2$,
$$
\frac{1}{h_1} + \frac{1}{h_2} = \frac{1}{H}
$$
De Pitagoras, la ecuacion a resolver es
$$
f(W) = \frac{1}{\sqrt{x_1^2 - W^2}} + \frac{1}{\sqrt{x_2^2 - W^2}} - \frac{1}{H} = 0
$$
Despejando $W$ del primer termino, se define $W_{k+1} = g(W_k)$
$$
W = g(W) = \sqrt{x_1^2 - \frac{H^2(x_2^2 - W^2)}{(\sqrt{x_2^2 - W^2}-H)^2}}
$$



```python
import numpy as np

#Constantes
x1 = 20
x2 = 30
H = 8

# Definicion de f(x) = 0
def f(w):

    return 1/(np.sqrt(x1**2 - w**2)) + 1/(np.sqrt(x2**2 - w**2)) - 1/H

def g(w):
    return np.sqrt(x1**2 - ((H**2*(x2**2 - w**2))/(np.sqrt(x2**2 - w**2)-H)**2))
    #return np.sqrt(x1**2 - ((H**2)*(x2**2-w**2))/(np.sqrt(x2**2 - w**2) - H**2))

# Parametros de inicializacion
p0 = 12
tol = 1e-5
n = 10

# %%  RUTINA006##########################
for i in range(n):
 p=g(p0)
 err=np.abs(p-p0)
 relerr=err/np.abs(p)
 if tol>err or tol>relerr or tol>np.abs(f(p)):
   break
 p0=p

print("Analisis de convergencia: ")
print("error absoluto: ", err)
print("error relativo: ", relerr)
print("valor f(p): ", np.abs(f(p)))
print("Cantidad de iteraciones: ", i+1)

print("-----------")
print("a) Cantidad de iteraciones: ",i+1)
print("b) El valor de W: ", np.around(p,6))
print("c) Error absoluto: ",np.around(err,6))
print("d) Error relativo: ",np.around(relerr,6))
```
