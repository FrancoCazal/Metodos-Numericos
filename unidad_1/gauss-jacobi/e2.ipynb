```python
import numpy as np

A = np.array([[35,12,15],
              [3,12,4],
              [7,12,26]], dtype=float)

B = np.array([[15140],
              [5710],
              [9570]], dtype=float)

P = np.array([[0],[0],[0]])

tol=1e-2
m=50
n=A.shape[0]
X=np.zeros((n,1), dtype=float)

# Rutina 2 - Jacobi
for i in range(m):
    for j in range(n):          
        X[j] = (B[j] - A[j, np.delete(np.arange(n), j) ].dot(P[np.delete(np.arange(n), j) ])) / A[j, j]
    err = np.linalg.norm(X - P)
    normX = np.linalg.norm(X)
    relerr = err / normX
    if tol > err or tol > relerr:
        break
    else:
        P = X.copy()

print(f"Los valores de respuesta son: \n{X} \ny la cantidad de iteraciones es {i+1}")
```
