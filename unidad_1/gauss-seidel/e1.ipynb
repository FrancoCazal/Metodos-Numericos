```python
import numpy as np

A = np.array([
    [21, 2, 5, 2, 5, 2],
    [5, 32, 6, 5, 6, 5],
    [1, 2, 12, 1, 2, 1],
    [1, 3, 1, 14, 3, 1],
    [3, 5, 3, 5, 24, 3],
    [3, 4, 3, 4, 3, 22]], dtype=float)

B = np.array([
    [109],
    [-149],
    [71],
    [-132],
    [215],
    [-115]], dtype=float)

C = np.hstack((A, B))

m=100
n=A.shape[0]
tol=1e-2
X=np.zeros([n,1],dtype=float)
P=np.zeros([n,1],dtype=float)

for i in range(m):
    for j in range(n):          
        X[j] = (B[j] - A[j, np.delete(np.arange(n), j) ].dot(X[np.delete(np.arange(n), j) ])) / A[j, j]
    err = np.linalg.norm(X - P)
    normX = np.linalg.norm(X)
    relerr = err / normX
    if tol > err or tol > relerr:
        break
    else:
        P = X.copy()

print(X)
```
