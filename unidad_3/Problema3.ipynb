## Problema 3 - Newton-Raphson multidimensional

### Consideraciones para la Resolución

* **Utilizar seis decimales para los resultados.**
* **Utilizar todos los decimales para los cálculos.**
* **No utilizar formato exponencial (notación científica).**
* **Utilizar una tolerancia de $10^{-8}$.**
* **No modificar los valores iniciales ni la tolerancia para los cálculos.**
* **No es necesario realizar ninguna conversión de unidades.**

Un objeto que cae verticalmente a través del aire está sujeto a una resistencia viscosa, así como a la fuerza de gravedad. Suponga que un objeto con masa $m$ (lb) cae desde una altura $s_0$ (ft) y que la altura del objeto después de $t$ (s) es:

$$s(t)=s_0-\frac{mg}{k}t+\frac{m^2g}{k^2}(1-e^{-\frac{kt}{m}})$$

donde $g$ (ft/s²) es la gravedad y $k$ (lb.s/ft) representa el coeficiente de la resistencia del aire.

Sabiendo que de acuerdo a las pruebas realizadas se obtuvieron los siguientes resultados:

| Masa ($m$) | Tiempo ($t$) | Altura ($s$) |
| :---: | :---: | :---: |
| 100 | 0.5 | 35.8762 |
| 60 | 0.7 | 31.9207 |
| 20 | 0.1 | 39.835 |

Utilizando el método de Newton para resolver el sistema, con una tolerancia de $10^{-8}$, y un valor inicial unitario para la iteración inicial.

**Determinar:**

* **Item a:** El valor de $s_0$
* **Item b:** El valor de $g$
* **Item c:** El valor de $k$
* **Item d:** Si los valores reales son $s_0=40$, $g=33$, $k=0.18$, determinar el maximo error cometido

_Planteamiento_

Definiendo $f(s_0, g, k) = s_0 -\frac{mg}{k}t+\frac{m^2g}{k^2}(1-e^{-\frac{kt}{m}}) - s_t = 0$ para cada experimento de la tabla resulta:

$$
f_1(s_0, g, k) = s_0 -\frac{100g}{k}0.5+\frac{100^2g}{k^2}(1-e^{-\frac{0.5k}{100}}) - 35.8762 = 0
$$
$$
f_2(s_0, g, k) = s_0 -\frac{60g}{k}0.7+\frac{60^2g}{k^2}(1-e^{-\frac{0.7k}{60}}) - 31.9207 = 0
$$
$$
f_3(s_0, g, k) = s_0 -\frac{20g}{k}0.1+\frac{20^2g}{k^2}(1-e^{-\frac{0.1k}{20}}) - 39.835 = 0
$$
Lo anterior constituye un sistema de ecuaciones no lineal resoluble con el metodo de Newton-Raphson

```python
import jax.numpy as jnp
from jax import jacfwd

# Sistema F(X) = 0_3
def sistema(var):
    s0,g,k=var
    f1 = s0 - (100*g/k)*0.5 + (jnp.pow(100,2)*g/jnp.pow(k,2))*(1-jnp.exp(-0.5*k/100)) - 35.8762
    f2 = s0 - (60*g/k)*0.7 + (jnp.pow(60,2)*g/jnp.pow(k,2))*(1-jnp.exp(-0.7*k/60)) - 31.9207
    f3 = s0 - (20*g/k)*0.1 + (jnp.pow(20,2)*g/jnp.pow(k,2))*(1-jnp.exp(-0.1*k/20)) - 39.835
    return jnp.array([f1,f2,f3],dtype=float)

jacob_sist=jacfwd(sistema)

# PARAMETROS DE INICIALIZACION
P0 = jnp.array([1.0, 1.0, 1.0], dtype=float)
tol = 1e-8
m = 10

################RUTINA009##########################
for i in range(m):
    F = sistema(P0)
    J = jacob_sist(P0)
    deltaP = jnp.linalg.solve(J, -F)
    P = P0 + deltaP
    err = jnp.linalg.norm(deltaP)
    relerr = err / jnp.linalg.norm(P)
    f_norm = jnp.linalg.norm(F)
    if (err < tol) or (relerr < tol) or (f_norm < tol):
        P0 = P
        break
    P0 = P

print("Analsis de convergencia:")
print("Error absoluto: ", err)
print("Error relativo: ", relerr)
print("Cantidad de iteraciones: ", i+1)
print("Valor de |F(P)|: ", f_norm)

s0_res, g_res, k_res = P

print("------------------")
print("a) Valor de S0: ", jnp.around(s0_res,6))
print("b) Valor de g: ", jnp.around(g_res,6))
print("c) Valor de k: ", jnp.around(k_res,6))
```
